```python
from faker import Faker
fake=Faker()
num_rows=1000
data={
    'staffid': range(51,num_rows+51),
    'staffname':[fake.name() for i in range(num_rows)],
    'Role':[fake.job() for i in range(num_rows)],
    'HireDate':[fake.date_between(start_date='-5y',end_date='today') for i in range(num_rows)],

}
stuff=pd.DataFrame(data)
stuff
```
